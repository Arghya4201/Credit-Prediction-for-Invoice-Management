# Milestone 3 - Feature Engineering and Selection


### 1. Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.


```python
Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace outliers with appropriate values
df['amount_in_usd'] = np.where((df['amount_in_usd'] < lower_bound) | (df['amount_in_usd'] > upper_bound),
                                df['amount_in_usd'].median(),
                                df['amount_in_usd'])

print(df['amount_in_usd'])
```

### 2. Label encoding or One hot Encoding on all the categorical columns 


```python
categorical = []
continuous = []

for col in df.columns:
    if df[col].dtype == 'object':
        categorical.append(col)
    else:
        continuous.append(col)

print("Categorical columns:")
print(categorical)

print("\nContinuous columns:")
print(continuous)
```

```python
le = LabelEncoder()
for col in categorical:
    if df[col].dtype == 'float' or df[col].dtype == 'int':
        df[col] = df[col].astype(str)
    df[col] = le.fit_transform(df[col].astype(str))

df.head()

```

### 3. Log Transformations on continuous columns 


```python
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].astype(str)
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].str.replace(',', '.')
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].str.replace('-', '')
df['RELEASED_CREDIT_VALUE'] = df['RELEASED_CREDIT_VALUE'].apply(lambda x: int(float(x)) if x != '0.00' else 0)

df['RELEASED_CREDIT_VALUE'] = (df["RELEASED_CREDIT_VALUE"]+1).transform(np.log10)
df['amount_in_usd'] = (df["amount_in_usd"]+1).transform(np.log10)
```

### 4. Try to extract new features by grouping existing columns 


```python
grouped_data = df.groupby(['COMPANY_CODE'])['ORDER_CURRENCY'].agg(['sum']).reset_index()
df = pd.merge(df, grouped_data, on='COMPANY_CODE', how='right')
df = df.drop('ORDER_CURRENCY', axis=1)
df = df.drop('COMPANY_CODE', axis=1)

```

### Note - All of you Introduce a particular module (Link Attached) for datetime manipulation.

```python
def difference_in_days(melt, lags, ffday, customer_id_col, create_date_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    
    melt = melt.reset_index(drop=True)
    
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Diff'] = melt.groupby([customer_id_col])['Last-'+str(i)+'day_Sales'].diff()
    melt = melt.fillna(0)
    return melt

lags = 7
ffday = 1
customer_id_col = 'unique_cust_id'
create_date_col = 'ORDER_CREATION_DATE'
net_amount_col = 'amount_in_usd'

df_date = difference_in_days(df, lags, ffday, customer_id_col, create_date_col, net_amount_col)
```
